```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/jessi/Downloads/CC GENERAL.csv")
df.head()
```

```python
df.drop('CUST_ID',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('CREDIT_LIMIT',axis=1,inplace=True)
df['MINIMUM_PAYMENTS']=df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean())
df.head()
```

```python
df.info()
```

```python
from sklearn.cluster import AgglomerativeClustering 
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)  
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df ['MINIMUM_PAYMENTS'] , df["BALANCE"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("MINIMUM_PAYMENTS")
ax.set_ylabel("BALANCE")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df, method="complete"))
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=4, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df)
df['label']=kmeans.labels_
print(kmeans.cluster_centers_)
```

```python
from sklearn.cluster import KMeans
Sum_of_squared_distances=[]
k=range(1,15)
for k in k:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.scatter(df["MINIMUM_PAYMENTS"][df['label'] == 0],          
            df["BALANCE"][df['label'] == 0],s=20,c='magenta',label='Careful')
plt.scatter(df["MINIMUM_PAYMENTS"][df['label'] == 1],
           df["BALANCE"][df['label'] == 1],s=20,c='yellow',label='Standard')
plt.scatter(df["MINIMUM_PAYMENTS"][df['label'] == 2],
           df["BALANCE"][df['label'] == 2],s=20,c='green',label='Target')
plt.scatter(df["MINIMUM_PAYMENTS"][df['label'] == 3], 
           df["BALANCE"][df['label'] == 3],s=20,c='cyan',label='Careless')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('MINIMUM_PAYMENTS')
plt.ylabel('BALANCE')
plt.legend()
plt.show()
```
